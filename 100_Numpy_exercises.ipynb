#### 44. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates (★★☆)

````python
a = np.random.uniform(1, 101, size = (10, 2))

r = np.sqrt(a[:, 0]**2 + a[:, 1]**2)
theta = np.arctan2(a[:, 1], a[:, 0])
a[:, 0] = r
a[:, 1] = theta
print(a)
````
